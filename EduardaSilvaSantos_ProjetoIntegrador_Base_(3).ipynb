```python
# === CARREGAMENTO ===
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx'

# === SEU CÓDIGO AQUI ===

# Carregue o dataset em uma variável chamada 'df'
df = pd.read_excel(url)

print("Formato do DataFrame df:", df.shape)
df.head()

```

```python
# === VALIDAÇÃO AUTOMÁTICA - NÃO MODIFICAR ===
def _validar_carregamento():
    """Valida se o dataset foi carregado corretamente."""
    try:
        # Verificar se df existe
        if 'df' not in globals():
            print("DataFrame 'df' não encontrado")
            return False

        _df = globals()['df']

        # Verificar tamanho mínimo
        if len(_df) <= 400000:
            print(f"Dataset muito pequeno: {len(_df):,} linhas (esperado: >400k)")
            return False

        # Verificar colunas essenciais
        _required = ['InvoiceNo', 'StockCode', 'Quantity', 'UnitPrice', 'CustomerID', 'Country']
        _missing = [c for c in _required if c not in _df.columns]

        if _missing:
            print(f"Colunas ausentes: {_missing}")
            return False

        print("VALIDAÇÃO PASSOU - Dataset carregado corretamente!")
        print(f"   - {len(_df):,} transações")
        print(f"   - {_df.shape[1]} colunas")
        return True

    except Exception as e:
        print(f"Erro na validação: {e}")
        return False

# Executar validação
_validar_carregamento()
```

```python
# === VALIDAÇÃO AUTOMÁTICA - NÃO MODIFICAR ===
def _validar_tipos():
    """Valida se InvoiceDate foi convertido para datetime."""
    try:
        if 'df' not in globals():
            print("DataFrame 'df' não encontrado")
            return False

        _df = globals()['df']
        _date_col = 'InvoiceDate'

        if _date_col not in _df.columns:
            print(f"Coluna '{_date_col}' não encontrada")
            return False

        _is_datetime = pd.api.types.is_datetime64_any_dtype(_df[_date_col])

        if not _is_datetime:
            print(f"'{_date_col}' deve ser datetime, atual: {_df[_date_col].dtype}")
            print("   DICA: Use pd.to_datetime(df['InvoiceDate'])")
            return False

        print("VALIDAÇÃO PASSOU - InvoiceDate é datetime!")
        print(f"   - Tipo: {_df[_date_col].dtype}")
        print(f"   - Range: {_df[_date_col].min()} a {_df[_date_col].max()}")
        return True

    except Exception as e:
        print(f"Erro: {e}")
        return False

_validar_tipos()
```

## 4.1 Limpeza (10 pts)


```python
# === TRATAMENTO DE DADOS FALTANTES ===

# 1. Análise de missing
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing por coluna:")
print(missing)

# 2. Decisão estratégica:
# - Remover registros sem CustomerID (não conseguimos analisar cliente)
# - Remover registros sem Description (produto desconhecido)

df = df.dropna(subset=['CustomerID', 'Description'])

# Opcional: garantir tipo inteiro para CustomerID
df['CustomerID'] = df['CustomerID'].astype(int)

# Conferir novamente
print("\nApós tratamento de missing:")
print(df.isnull().sum())


```

```python
# === TRATAMENTO DE OUTLIERS E INCONSISTÊNCIAS ===

# 1. Visão geral dos valores de Quantity e UnitPrice
print("Resumo Quantity:")
print(df['Quantity'].describe())
print("\nResumo UnitPrice:")
print(df['UnitPrice'].describe())

# 2. Remover transações com Quantity = 0
df = df[df['Quantity'] != 0]

# 3. Remover UnitPrice <= 0 (preço negativo ou zero não faz sentido para receita)
df = df[df['UnitPrice'] > 0]

# 4. Tratar outliers extremos com base em quantis (ex.: acima do 99º percentil)
q_quant = df['Quantity'].quantile(0.99)
q_price = df['UnitPrice'].quantile(0.99)

df = df[(df['Quantity'].abs() <= q_quant) & (df['UnitPrice'] <= q_price)]

print("\nApós tratamento de outliers:")
print(df['Quantity'].describe())
print(df['UnitPrice'].describe())
print("Quantidade de linhas após limpeza:", len(df))

```

Observações (outliers e inconsistências):
A análise descritiva de Quantity e UnitPrice revelou a presença de quantidades negativas e preços muito elevados, além de transações com quantidade igual a zero e preços menores ou iguais a zero. Quantidade zero e preços não positivos não representam vendas válidas, por isso esses registros foram removidos.
Em seguida, utilizei o 99º percentil para limitar valores extremos de quantidade (em valor absoluto) e de preço unitário, reduzindo o impacto de erros de cadastro ou casos raríssimos que distorceriam médias e gráficos. Mantive quantidades negativas moderadas, pois elas representam devoluções/cancelamentos legítimos, importantes para monitorar estornos e qualidade do portfólio. Após o tratamento, a distribuição de Quantity e UnitPrice fica mais estável e condizente com a realidade operacional.

```python
# === CRIAR DF_CLEAN ===
df_clean = df.copy()  # Modifique conforme necessário

print(f"✓ df_clean: {df_clean.shape}")
```

```python
# === VALIDAÇÃO AUTOMÁTICA - NÃO MODIFICAR ===
def _validar_limpeza():
    """Valida se df_clean foi criado adequadamente."""
    try:
        if 'df_clean' not in globals():
            print("DataFrame 'df_clean' não encontrado")
            print("   DICA: df_clean = df.copy() após sua limpeza")
            return False

        _df_clean = globals()['df_clean']
        _df = globals().get('df', None)

        if not isinstance(_df_clean, pd.DataFrame):
            print("df_clean deve ser DataFrame")
            return False

        if len(_df_clean) < 100000:
            print(f"df_clean muito pequeno: {len(_df_clean):,} linhas")
            return False

        _null_pct = _df_clean.isnull().sum().sum() / (len(_df_clean) * _df_clean.shape[1]) * 100

        print("VALIDAÇÃO PASSOU - df_clean criado!")
        if _df is not None:
            print(f"   - {len(_df_clean):,} linhas ({len(_df_clean)/len(_df)*100:.1f}% do original)")
        else:
            print(f"   - {len(_df_clean):,} linhas")
        print(f"   - Missing: {_null_pct:.2f}%")

        if _null_pct > 30:
            print("   AVISO: Alto percentual de missing (considere tratar)")

        return True

    except Exception as e:
        print(f"Erro: {e}")
        return False

_validar_limpeza()
```

## 4.2 Feature Engineering (10 pts)


```python
# === FEATURE: REVENUE ===
# OBRIGATÓRIO: Nome exato 'Revenue'

# === SEU CÓDIGO AQUI ===
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']

df_clean['Revenue'].describe()



```

```python
# === FEATURES TEMPORAIS ===
# OBRIGATÓRIO: Year, Month, DayOfWeek, Hour

# === SEU CÓDIGO AQUI ===
df_clean['Year'] = df_clean['InvoiceDate'].dt.year
df_clean['Month'] = df_clean['InvoiceDate'].dt.month
df_clean['DayOfWeek'] = df_clean['InvoiceDate'].dt.dayofweek  # 0=Seg
df_clean['Hour'] = df_clean['InvoiceDate'].dt.hour



```

```python
# === FEATURE: IS_CANCELLATION ===
# OBRIGATÓRIO: Nome exato 'Is_Cancellation'

# === SEU CÓDIGO AQUI ===
df_clean['Is_Cancellation'] = df_clean['InvoiceNo'].astype(str).str.startswith('C').astype(int)

df_clean['Is_Cancellation'].value_counts()


```

```python
# === FEATURES CRIATIVAS ===
# Crie pelo menos 3 features adicionais

# === SEU CÓDIGO AQUI ===

# 1. Flag para Reino Unido
df_clean['Is_UK'] = (df_clean['Country'] == 'United Kingdom').astype(int)

# 2. Semana do ano
df_clean['WeekOfYear'] = df_clean['InvoiceDate'].dt.isocalendar().week.astype(int)

# 3. Receita total por cliente (feature agregada com transform)
df_clean['Customer_TotalRevenue'] = df_clean.groupby('CustomerID')['Revenue'].transform('sum')

# 4. Número de faturas por cliente
df_clean['Customer_NumInvoices'] = df_clean.groupby('CustomerID')['InvoiceNo'].transform('nunique')



```

```python
# === CRIAR DF_FEATURES ===
df_features = df_clean.copy()

print(f"✓ df_features: {df_features.shape}")

```

Observações:
As features derivadas foram construídas para conectar o comportamento transacional às perguntas de negócio. Revenue consolida o valor monetário de cada linha de venda, permitindo análises de faturamento por produto, cliente, país e tempo. As variáveis temporais (Year, Month, DayOfWeek, Hour) viabilizam a análise de sazonalidade e padrões de consumo ao longo do calendário.
A feature binária Is_Cancellation identifica notas de devolução, suportando o monitoramento de cancelamentos e seu impacto na receita. Já as features criativas (Is_UK, WeekOfYear, Customer_TotalRevenue, Customer_NumInvoices) ajudam a medir a dependência do negócio do mercado britânico, capturar sazonalidade semanal e diferenciar clientes de maior valor em termos de receita total e recorrência de compras.

```python
# === VALIDAÇÃO AUTOMÁTICA - NÃO MODIFICAR ===
def _validar_features():
    """Valida se features obrigatórias foram criadas."""
    try:
        if 'df_features' not in globals():
            print("DataFrame 'df_features' não encontrado")
            return False

        _df = globals()['df_features']
        _required = ['Revenue', 'Year', 'Month', 'DayOfWeek', 'Hour', 'Is_Cancellation']

        _missing = [f for f in _required if f not in _df.columns]
        _present = [f for f in _required if f in _df.columns]

        print(f"Features obrigatórias: {len(_present)}/{len(_required)}")

        for feat in _present:
            print(f"   {feat}")

        for feat in _missing:
            print(f"   {feat} - AUSENTE!")

        # Validações específicas
        if 'Revenue' in _df.columns:
            if _df['Revenue'].sum() <= 0:
                print("   Revenue deve ter valores positivos")

        if 'Is_Cancellation' in _df.columns:
            _unique = _df['Is_Cancellation'].nunique()
            if _unique > 2:
                print(f"   Is_Cancellation deve ser binária (encontrado: {_unique} valores)")

        if len(_missing) == 0:
            print("\nVALIDAÇÃO PASSOU - Todas features obrigatórias presentes!")
            return True
        else:
            print(f"\nATENÇÃO: {len(_missing)} feature(s) ausente(s)!")
            return False

    except Exception as e:
        print(f"Erro: {e}")
        return False

_validar_features()
```

## 5.1 Análise Temporal OBRIGATÓRIA (10 pts)
**Requisitos:**
- Agregação temporal
- Médias móveis (7 e 30 dias)
- Decomposição temporal
- Detecção de anomalias

```python
# === AGREGAÇÃO TEMPORAL ===

# Agregar receita por dia
df_temporal = (df_features
               .set_index('InvoiceDate')
               .resample('D')
               .agg({'Revenue': 'sum'}))

# Remover dias sem movimento, se quiser
df_temporal = df_temporal[df_temporal['Revenue'] > 0]

df_temporal.head()


```

```python
# === MÉDIAS MÓVEIS ===

df_temporal['MA_7'] = df_temporal['Revenue'].rolling(window=7, min_periods=1).mean()
df_temporal['MA_30'] = df_temporal['Revenue'].rolling(window=30, min_periods=1).mean()

plt.figure()
plt.plot(df_temporal.index, df_temporal['Revenue'], label='Revenue diário', alpha=0.4)
plt.plot(df_temporal.index, df_temporal['MA_7'], label='Média móvel 7 dias')
plt.plot(df_temporal.index, df_temporal['MA_30'], label='Média móvel 30 dias')
plt.title('Revenue diário e médias móveis')
plt.xlabel('Data')
plt.ylabel('Revenue')
plt.legend()





```

```python
# === DECOMPOSIÇÃO TEMPORAL ===
# Separe a série em: Tendência + Sazonalidade + Resíduo

from statsmodels.tsa.seasonal import seasonal_decompose

serie = df_temporal['Revenue']

# Decomposição com período semanal (7 dias)
result = seasonal_decompose(serie, model='additive', period=7)

fig = result.plot()
fig.set_size_inches(10, 8)
plt.suptitle('Decomposição da série de Revenue diário', y=1.02)





```

```python
# === DETECÇÃO DE ANOMALIAS ===
# Identifique dias com receita ANORMALMENTE alta ou baixa
# Interprete os resultados (DOCUMENTE)
# === DETECÇÃO DE ANOMALIAS ===
# Identificação de dias com receita anormalmente alta ou baixa

media = df_temporal['Revenue'].mean()
desvio = df_temporal['Revenue'].std()

df_temporal['Z_Revenue'] = (df_temporal['Revenue'] - media) / desvio

# critério simples: |Z| > 3
anomalias = df_temporal[np.abs(df_temporal['Z_Revenue']) > 3]

print("Dias anômalos (|Z| > 3):")
display(anomalias.head())

plt.figure()
plt.plot(df_temporal.index, df_temporal['Revenue'], label='Revenue diário', alpha=0.5)
plt.scatter(anomalias.index, anomalias['Revenue'], label='Anomalias', marker='o')
plt.title('Detecção de anomalias na receita diária')
plt.xlabel('Data')
plt.ylabel('Revenue')
plt.legend()
```

```python
# === CRESCIMENTO MOM (Month-over-Month) ===

monthly_rev = (df_features
               .groupby(['Year', 'Month'])['Revenue']
               .sum()
               .reset_index())

monthly_rev['YearMonth'] = pd.to_datetime(dict(
    year=monthly_rev['Year'],
    month=monthly_rev['Month'],
    day=1
))

monthly_rev = monthly_rev.sort_values('YearMonth')
monthly_rev['MoM_Growth'] = monthly_rev['Revenue'].pct_change() * 100

plt.figure()
plt.plot(monthly_rev['YearMonth'], monthly_rev['Revenue'], marker='o')
plt.title('Revenue mensal')
plt.xlabel('Ano-Mês')
plt.ylabel('Revenue total')
plt.xticks(rotation=45)

plt.figure()
plt.bar(monthly_rev['YearMonth'].astype(str), monthly_rev['MoM_Growth'])
plt.title('Crescimento MoM (%)')
plt.xlabel('Ano-Mês')
plt.ylabel('Crescimento %')
plt.xticks(rotation=45)


```

Observações:
A série de receita diária apresenta forte variação, com picos concentrados em períodos específicos, o que é esperado em um negócio de presentes com sazonalidade comercial. As médias móveis de 7 e 30 dias suavizam o ruído de curto prazo e evidenciam uma tendência geral de aumento ao longo do período analisado.
A decomposição temporal indica a presença de um componente sazonal consistente, associado a padrões semanais de compra, além de flutuações residuais que podem estar ligadas a campanhas pontuais ou eventos específicos. A análise de anomalias mostra alguns dias com receita significativamente acima ou abaixo do padrão histórico, que merecem investigação operacional e comercial mais detalhada.
A comparação mês a mês revela variações relevantes na receita, sugerindo que determinados meses concentram oportunidade maior de faturamento e outros representam pontos de atenção para ações corretivas ou de estímulo de demanda.
